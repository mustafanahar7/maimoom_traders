```python
query = 'SELECT * FROM mainapp_customers_order_from_website'  # Replace with your actual table name
df = pd.read_sql(query, engine)
df

# Preprocessing
df['created_at'] = pd.to_datetime(df['created_at'])  # Convert created_at column to datetime
df = df.set_index('created_at')

# Feature engineering
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month
df['quarter'] = df.index.quarter

# Aggregating total sales by day
daily_sales = df.resample('D').sum()['total_amount']

# Creating a DataFrame for modeling
df_model = daily_sales.to_frame().reset_index()
df_model['day_of_week'] = df_model['created_at'].dt.dayofweek
df_model['month'] = df_model['created_at'].dt.month
df_model['quarter'] = df_model['created_at'].dt.quarter
df_model['lag_1'] = df_model['total_amount'].shift(1)
df_model['lag_7'] = df_model['total_amount'].shift(7)
df_model = df_model.dropna()

# Features and target variable
X = df_model.drop(columns=['created_at', 'total_amount'])
y = df_model['total_amount']

# Split the data with a fixed random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# Initialize and train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Make predictions
lr_y_pred = lr_model.predict(X_test)

# Evaluate the model
lr_mae = mean_absolute_error(y_test, lr_y_pred)
lr_rmse = mean_squared_error(y_test, lr_y_pred, squared=False)
lr_r2 = r2_score(y_test, lr_y_pred)
lr_accuracy_percentage = lr_r2 * 100

print(f'Linear Regression MAE: {lr_mae}')
print(f'Linear Regression RMSE: {lr_rmse}')
print(f'Linear Regression R Score: {lr_r2}')
print(f'Linear Regression Accuracy: {lr_accuracy_percentage:.2f}%')
```

```python
import pandas as pd
from sqlalchemy import create_engine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define parameter grid
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the model
rf_model = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV
rf_grid_search = GridSearchCV(estimator=rf_model, param_grid=rf_param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the model
rf_grid_search.fit(X_train, y_train)

# Best parameters
print(rf_grid_search.best_params_)

# Make predictions with the best estimator
best_rf_model = rf_grid_search.best_estimator_
rf_y_pred = best_rf_model.predict(X_test)

# Evaluate the model
rf_mae = mean_absolute_error(y_test, rf_y_pred)
rf_rmse = mean_squared_error(y_test, rf_y_pred, squared=False)
rf_r2 = r2_score(y_test, rf_y_pred)
rf_accuracy_percentage = rf_r2 * 100

print(f'Random Forest MAE: {rf_mae}')
print(f'Random Forest RMSE: {rf_rmse}')
print(f'Random Forest R Score: {rf_r2}')
print(f'Random Forest Accuracy: {rf_accuracy_percentage:.2f}%')
```
